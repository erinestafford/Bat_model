```python
import numpy as np
import random
import pickle
import os
import matplotlib.pyplot as plt
```

```python
#['PMP', 'PSA', 'RDA', 'FAV', 'TAV'] is 
#[movement probabilities between patches, foraging switches per night, roost duration,foraging avg visitors,total avg visitors]
files = [f for f in os.listdir() if f[0]=='o']
out_pmp_roost = np.zeros((6,101))
out_pmp_residential = np.zeros((6,101))
out_pmp_orchard = np.zeros((6,101))
out_pmp_water_body = np.zeros((6,101))
out_pmp_forest = np.zeros((6,101))
out_pmp_dump = np.zeros((6,101))
out_psa = np.zeros(101)
out_rda = np.zeros(101)
out_fav = np.zeros((900,101))
out_tav = np.zeros((900,101))
count =0
for f in files:
    with open(f, 'rb') as out:
        test_out = pickle.load(out)
        out_pmp_roost[:,count]=test_out['PMP']['Roost']
        out_pmp_residential[:,count]=test_out['PMP']['Residential']
        out_pmp_orchard[:,count]=test_out['PMP']['Orchard']
        out_pmp_water_body[:,count]=test_out['PMP']['Water Body']
        out_pmp_forest[:,count]=test_out['PMP']['Forest']
        out_pmp_dump[:,count]=test_out['PMP']['Dump']
        out_psa[count]=np.mean(test_out['PSA'])
        out_rda[count]=np.mean(test_out['RDA'])
        out_fav[:,count] = test_out['FAV']
        out_tav[:,count] = test_out['TAV']
        count+=1
```

```python
plt.boxplot(out_psa)
plt.show()
```

```python
plt.boxplot(out_rda)
plt.show()
```

```python
plt.imshow(out_fav[:,0].reshape(30,30))
plt.colorbar()
```
