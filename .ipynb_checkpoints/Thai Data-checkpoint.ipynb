```python
import sys
import csv
import numpy as np
import matplotlib.pyplot as plt
data=[]
with open('Thai_LU_data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        spamreader2 = csv.reader(row, delimiter=',', quotechar='|')
        for num in spamreader2:
            data.append(num)
```

```python
data2=data[1:] 
```

```python
data3 = np.array(data2)
```

```python
thai_data = np.zeros(data4.shape)
for i in range(len(data4)):
    for j in range(len(data4[i])):
        thai_data[i,j]= int(data4[i,j])
```

```python
#[Name, K_min (g),K_max (g),nu_min /hr,nu_max /hr]
resources_map = { 10: ['Orchard', 350*100, 350*1000,0.05,0.1],
             11: ['Forest', 200,2000,0.0,0.1],
             12: ['Forest', 200,2000,0.0,0.1],
             20: ['Orchard',350*100, 350*1000,0.05,0.1],
             50: ['Forest', 200,2000,0.0,0.1],
             60: ['Forest', 200,2000,0.0,0.1],
             61: ['Forest', 200,2000,0.0,0.1],
             62: ['Forest', 200,2000,0.0,0.1],
             70: ['Forest', 200,2000,0.0,0.1],
             71: ['Forest', 200,2000,0.0,0.1],
             72: ['Forest', 200,2000,0.0,0.1],
             80: ['Forest', 200,2000,0.0,0.1],
             81: ['Forest', 200,2000,0.0,0.1],
             82: ['Forest', 200,2000,0.0,0.1],
             90: ['Forest', 200,2000,0.0,0.1],
             120: ['Forest',200,2000,0.0,0.1],
             121: ['Forest',200,2000,0.0,0.1],
             122: ['Forest',200,2000,0.0,0.1],
             130: ['Forest',200,2000,0.0,0.1],
             140: ['Forest',200,2000,0.0,0.1],
             150: ['Forest',200,2000,0.0,0.1],
             152: ['Forest',200,2000,0.0,0.1],
             153: ['Forest',200,2000,0.0,0.1],
             180: ['Forest',200,2000,0.0,0.1],
             190: ['Residential',200,1000,0.1,0.5],
             200: ['Dump', 200,1000,0.1,0.5],
             201: ['Dump', 200,1000,0.1,0.5],
             202: ['Dump', 200,1000,0.1,0.5],
             210: ['Water Body',200,3000,0.0,0.1],
             220: ['Water Body',200,3000,0.0,0.1],
             250: ['Residential',200,3000,0.1,0.5]
}
```

```python
patch_net = {}
patch_names = []
resources = np.zeros(thai_data.shape)
count = 0
for i in range(thai_data.shape[0]):
    for j in range(thai_data.shape[1]):
        patch_net[count] ={}
        patch_net[count]['patch_center']= [i+0.015,j+0.015]
        patch_net[count]['Res']=np.array([0.5,0.5])
        attr = resources_map[int(thai_data[i][j])]
        K = np.random.choice(np.arange(attr[1],attr[2]))
        patch_net[count]['carrying_capacity']= K
        patch_net[count]['init_resources']= np.random.choice(np.arange(0,K))
        patch_net[count]['resource_birth'] = np.random.choice(np.arange(attr[3],attr[4],0.001))
        patch_net[count]['Name'] = attr[0]
        patch_names.append(patch_net[count]['Name'])
        resources[i][j] = patch_net[count]['init_resources']
        count = count + 1
```

```python
forest_patches = []
for p in range(len(patch_names)):
    if patch_names[p] in 'Forest':
        forest_patches.append(p)
```

```python
import random
roost=random.choice(forest_patches)
patch_net[roost]['Name']='Roost'
```

```python
import pickle 

with open('Thai_net.pkl', 'wb') as f:
    pickle.dump(patch_net, f)
```

```python
data=np.array(resources)
plt.imshow(data,cmap='magma')
plt.colorbar()
plt.show()
```

```python
import rasterio
from rasterio.enums import Resampling

upscale_factor = 30/1000

with rasterio.open('test.tif') as dataset:

    # resample data to target shape
    data = dataset.read(
        out_shape=(
            dataset.count,
            int(dataset.height * upscale_factor),
            int(dataset.width * upscale_factor)
        ),
        resampling=Resampling.nearest
    )

    # scale image transform
    transform = dataset.transform * dataset.transform.scale(
        (dataset.width / data.shape[-1]),
        (dataset.height / data.shape[-2])
    )
    show(data,cmap="magma",transform=transform)
```

```python
data = data.reshape(60,60)
```

```python
patch_net = {}
patch_names = []
resources = np.zeros(data.shape)
count = 0
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        patch_net[count] ={}
        patch_net[count]['patch_center']= [i+0.015,j+0.015]
        patch_net[count]['Res']=np.array([0.5,0.5])
        attr = resources_map[int(data[i][j])]
        K = np.random.choice(np.arange(attr[1],attr[2]))
        patch_net[count]['carrying_capacity']= K
        patch_net[count]['init_resources']= np.random.choice(np.arange(0,K))
        patch_net[count]['resource_birth'] = np.random.choice(np.arange(attr[3],attr[4],0.001))
        patch_net[count]['Name'] = attr[0]
        patch_names.append(patch_net[count]['Name'])
        resources[i][j] = patch_net[count]['init_resources']
        count = count + 1
```

```python
forest_patches = []
for p in range(len(patch_names)):
    if patch_names[p] in 'Forest':
        forest_patches.append(p)
```

```python
import random
roost=random.choice(forest_patches)
patch_net[roost]['Name']='Roost'
```

```python
import pickle 

with open('Thai_net.pkl', 'wb') as f:
    pickle.dump(patch_net, f)
```
